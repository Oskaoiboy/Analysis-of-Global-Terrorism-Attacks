# Analysis of Global Terrorism Attacks

## 1. Introduction: Global Terrorism

Terrorism is, in the broadest sense, the use of intentional violence for political or religious purposes. It is used in this regard primarily to refer to violence during peacetime or in the context of war against non-combatants (mostly civilians and neutral military personnel).

Information on over 180,000 terrorist attacks There are 135 columns including Event, date, time, country, region, place, killed, wounded, if there was a ransom, the result if there was a suicide attack, used weapon.. The Global Terrorism Database (GTD) is an open source database of terrorist attacks worldwide from 1970 to 2017. GTD includes systematic data on domestic and international terrorist incidents that occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and the Response to Terrorism (START), headquartered at the University of Maryland.

The purpose of this analysis is to study terrorist events around the world. I analyze murders, wounded, terrorist group activity in different years and places, property damage. I also analyze the group of criminals and the nationality of the criminals, the geographic region where the attack took place, where the criminals escaped (the country or city they asked to go after the attack), the weapon used in the attack.

Data Dictionary:

'Year' : year of the incident (i.e. 1970, 2002, etc.)

'Month' : month of the incident (i.e. 1, 2, 4, 12, etc.)

'Day' : day of the incident  (i.e. 1, 31, 15, 7, etc.)

'Country' : country where the incident occurred  (i.e. Mexico, United States, etc.)

'Region' : region where the incident occurred   (i.e. South Asia, Middle East & North Africa, etc.)

'city' : city where the incident occurred     (i.e. Fukouka, Cairo, etc.)
 
'latitude' :  latitude where the incident occurred   (i.e. 18.456792, 19.371887, etc.)

'longitude' : longitude where the incident occurred   (i.e. -69.951164, -99.086624, etc.)

'Attack Type' : general method of attack used (i.e. assassination, hijacking, bombing/explosion, etc.)

'Killed' : number of dead   (i.e. 20.0, 55.0, etc.)

'Wounded' : number of wounded  (i.e. 15.0, 322.0, etc.)

'Target' : general type of target/victim (i.e. business, government, police, military, etc.)

'Summary' : a short description of what happened  (i.e.  Two gunmen shot and killed Egyptian , 322.0, etc.)

'Group' : Terrorist group    (i.e.  Black Nationalists, Tupamaros (Uruguay), etc.)
 
'Target Type' : specific person, building, installation, etc. that was targeted "

'Weapon Type' : general type of weapon used in the incident (i.e. biological, chemical, firearms, etc.)

'Motive' :  Attack motive (i.e. To protest the Cairo Illinois Police Deparment	etc.)

'casualities' : 'Killed' + 'Wounded'  (i.e.  144.0, 142.0, etc.)

# 4. Data acquisiton: scrapping data from website 
## (https://tradingeconomics.com/country-list/terrorism-index)

```python
import pandas as pd        
import requests #Get contents from the web
from bs4 import BeautifulSoup #Extract information
import csv #The so-called CSV (Comma Separated Values) format is the most common import and export format for spreadsheets and databases. 

text1 = []                   #A box bracket pair denotes one of a list object, or an index subscript,
text2 = []
text3 = []
text4 = []

req = requests.get("https://tradingeconomics.com/country-list/terrorism-index") #Requesting data from site
soup = BeautifulSoup(req.content,'lxml') #BeautifulSoup represents the document as a nested data structure:
textof1 = soup.find_all("tr", class_="datatable-row") #Searching all "tr" with class "datatable-row"
for x in textof1:       #loop
    text = x.find_all("td") 
    text1.append(text[0].text.strip()) #append = Add an data to the list; 
    text2.append(text[1].text.strip()) #strip() method = removes any leading and trailing characters
    text3.append(text[2].text.strip())
    text4.append(text[3].text.strip())
textof1 = soup.find_all("tr", class_="datatable-row-alternating") #Searching all "tr" with class "datatable-row-alternating"
for x in textof1:
    text = x.find_all("td")
    text1.append(text[0].text.strip())
    text2.append(text[1].text.strip())
    text3.append(text[2].text.strip())
    text4.append(text[3].text.strip())

alli = {'Country': text1, 'Last': text2, 'Previous': text3, 'Reference':text4}
print(alli)
df = pd.DataFrame(alli)
df.to_csv('tradingeconomics.csv')      #save all data to "tradingeconomics.csv"
```

## 5.3  Analyze who suffers more from terrorist attacks.

In this section we are analyzing who suffers more from terrorist attacks.

```python
def plot_Most_Affected_Target(kind='Region',big=(30,20)):
    sns.set(style="white",font_scale=2.5)
    fig = plt.figure(figsize=big)
    ax1 = fig.add_subplot(3,2,1)
    ax1.set_title('Most Affected Target')
    ax1.set_ylabel('');
    data[data['Year']>=2017]['EventID'].groupby(data[kind]).count().sort_values(ascending=False)[:10].plot(kind='barh');
    ax1.set_ylabel('');
plot_Most_Affected_Target(kind='Target_type',big=(30,30));
```

Citizens & Private Property, Military, Police and Business are mostly targeted, but mainly citizens, then we see a steep decline in frequency for other types which then gradually declines. In the count tile by Types, naturally Bombing/Explosion has the brightest cases, but this tile also shows where this and other types happen the most.
